```python
import pandas as pd

data = pd.read_csv("https://media.geeksforgeeks.org/wp-content/uploads/tips.csv")

data.head(10)
```

## matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Scatter plot with day against tip
plt.scatter(data['day'], data['tip'])
 
# Adding Title to the Plot
plt.title("Scatter Plot")
 
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
 
plt.show()
```

```python
%matplotlib inline

import ipywidgets as widgets
import numpy as np

o1 = widgets.Output()
o2 = widgets.Output()
d1 = pd.DataFrame(np.random.normal(size = 50))
d2 = pd.DataFrame(np.random.normal(size = 100))

tab = widgets.Tab(children = [o1, o2])
tab.set_title(0, "First")
tab.set_title(1, "Second")
display(tab)

with o1:
    fig1, axes1 = plt.subplots()
    d1.hist(ax = axes1)
    plt.show(fig1)


with o2:
    fig2, axes2 = plt.subplots()
    d2.hist(ax = axes2)
    plt.show(fig2)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

 
# Scatter plot with day against tip
plt.scatter(data['day'], data['tip'], c=data['size'],
            s=data['total_bill'])
 
# Adding Title to the Plot
plt.title("Scatter Plot")
 
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
 
plt.colorbar()
 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

 
# Scatter plot with day against tip
plt.plot(data['tip'])
plt.plot(data['size'])
 
# Adding Title to the Plot
plt.title("Scatter Plot")
 
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

 
# Bar chart with day against tip
plt.bar(data['day'], data['tip'])
 
plt.title("Bar Chart")
 
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
 
# Adding the legends
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
 
# histogram of total_bills
plt.hist(data['total_bill'])
 
plt.title("Histogram")
 
# Adding the legends
plt.show()
```

## Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# draw lineplot
sns.lineplot(x="sex", y="total_bill", data=data)
 
# setting the title using Matplotlib
plt.title('Title using Matplotlib Function')
 
plt.show()
```

### Scatter Plot

```python
sns.scatterplot(x='day', y='tip', data=data,)
plt.show()
```

```python
sns.scatterplot(x='day', y='tip', data=data,
               hue='sex')
plt.show()
```

### Line Plot

```python
sns.lineplot(x='day', y='tip', data=data)
plt.show()
```

```python
# using only data attribute
sns.lineplot(data=data.drop(['total_bill'], axis=1))
plt.show()
```

### Bar Plot

```python
sns.barplot(x='day',y='tip', data=data,
            hue='sex')
 
plt.show()
```

### Histogram

```python
sns.histplot(x='total_bill', data=data, kde=True, hue='sex')
 
plt.show()
```

## Bokeh

```python
# importing the modules
from bokeh.plotting import figure, output_notebook, show
from bokeh.palettes import magma
import pandas as pd

output_notebook()
```

### Scatter Plot

```python
# instantiating the figure object
graph = figure(title = "Bokeh Scatter Graph")
 
color = magma(256)
 
# plotting the graph

graph.scatter(data['total_bill'], data['tip'], color=color)
 
# displaying the model
show(graph)
```
